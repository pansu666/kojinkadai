```python
#ライブラリ読み込み
import pandas as pd
import numpy as np
import re  #正規式
import demoji   #顔文字
import neologdn  #テキスト正規化
```

# ファイル読み込み

```python
DATA_IN_PATH="C:/r_analysis/1.20221101/"
zenshori_data = pd.read_csv(DATA_IN_PATH + 'cleansing.csv',encoding="cp932" , header=0)
zenshori_data.head()
```

# レビューデータのリスト化
## 顔文字除去・全角文字→半角文字変換

```python
review_list=list()

for i in zenshori_data['review']:
    a = demoji.replace(string=i, repl='')
    a = neologdn.normalize(a)
    review_list.append(a)
    
len(review_list)  #確認
```

```python
#サンプル取って確認
review_list[19273:19290]
```

```python
#確認
zenshori_data['review']==review_list
```

# to csv

```python
review_list=pd.DataFrame(review_list)  #to dataFrame
zenshori_data['review_cleansing']=review_list
zenshori_data=zenshori_data[['Unnamed: 0','name','date','sex','retire','type','label_k','review_cleansing']]
zenshori_data.columns=['no','name','date','sex','retire','type','label','review_cleansing']
zenshori_data   #data 確認
```

```python
#save data
DATA_OUT_PATH = 'C:/r_analysis/1.20221101/'
zenshori_data.to_csv(DATA_OUT_PATH + 'train_cleansing.csv', index=False, quoting=3, encoding="CP932") 
```
